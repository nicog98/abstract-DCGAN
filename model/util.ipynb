```python
# get all image files in directory specified by from_path and store resized images into to_path
def resize_and_save_images(from_path, to_path):
    for filename in listdir(from_path):
        # skip hidden files
        if filename[0] == ".":
            continue
        # resize image
        image = Image.open(from_path + '/' + filename)
        image_resized = image.resize((64, 64))
        # save in new directory
        image_resized.save(to_path + '/' + filename)
```

```python
# to be used for celebA images dataset
def crop_and_resize_images(from_path, to_path):
    for filename in listdir(from_path):
        if filename.endswith('.jpg'):
            image = Image.open(from_path + '/' + filename)
            x1 = 30
            y1 = 40
            image_cropped = image.crop((x1,y1,x1+138,y1+138))
            image_resized = image.resize((64,64))
            image_resized.save(to_path + '/' + filename)
```

```python
# augment training data
# rotate and flip each image
def augment_image_data(data_path):
    for filename in listdir(data_path):
        # skip hidden files
        if filename[0] == ".":
            continue
        image = Image.open(data_path + '/' + filename)
        # flip image horizontally
        hoz_flip_image = image.transpose(Image.FLIP_LEFT_RIGHT)
        # flip image vertically
        vert_flip_image = image.transpose(Image.FLIP_TOP_BOTTOM)
        # rotate image
        rotate_image = image.rotate(90)
        # save images
        hoz_flip_image.save(data_path + '/' +  "_00.".join(filename.split(".")))
        vert_flip_image.save(data_path + '/' + "_01.".join(filename.split(".")))
        rotate_image.save(data_path + '/' + "_02.".join(filename.split(".")))
```

```python
augment_image_data(save_path)
```
